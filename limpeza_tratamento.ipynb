```python
# Verificando a correlação entre o preço do plano e a quantidade vendida e listando os planos que mais venderam
# priemiro vamos extrair as vendas concluidas do dataset

Vendas = Dados_Teste.loc[(Dados_Teste['Status'] == "Ganho")]
Vendas
```

```python
# segundo: vamos criar um novo dataframe agrupando as variaveis de plano para entender quantidade vendida por plano e preço médio
# vamos tirar uma média do preço já que o preço é diferente mesmo em negociações de mesmo plano


DadosCor = Vendas.groupby(['Plano', 'Status']).agg({'Valor': 'median', 'Plano': 'count'})
DadosCor = DadosCor.rename(columns={'Plano': 'QtdePlano'})
# agora vamos ordenar os planos por quantidade vendida para saber qual mais vende
DadosCor = DadosCor.sort_values(by='QtdePlano', ascending=False)
DadosCor

# o plano que mais vendeu foi o Essencial com uma mediana de preço de 1.990 reais
```

```python
#Verificando a correlação entre o preço do plano e a quantidade vendida
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.implot(x = 'Valor', y = "Status", data = Dados_Teste);


plt.scatter(x= DadosCor.QtdePlano, y= DadosCor.Valor, color = 'blue', facecolors = 'none', marker = '.')
plt.title('Correlacao')
plt.xlabel('QtdePlano')
plt.ylabel('Valor')
plt.show()


# podemos ver que embora tenhamos vendas como outliers, temos um correlação entre o preço e o a quantidade vendida
# os planos abaixo de 3 mil reais foram os mais vendidos
```

```python
# aqui determinamos a média, mediana e moda
media_vendas = Dados_Teste[Dados_Teste['Status'] == 'Ganho']['Valor'].mean()
mediana_vendas = Dados_Teste[Dados_Teste['Status'] == 'Ganho']['Valor'].median()
moda_vendas = Dados_Teste[Dados_Teste['Status'] == 'Ganho']['Valor'].mode()

print("Média para vendas com Status 'Ganho':", media_vendas)
print("Mediana para vendas com Status 'Ganho':", mediana_vendas)
print("Moda para vendas com Status 'Ganho':", moda_vendas[0])
```
